```python
# 透過yfinance抓取過去一年台積電股價
import yfinance as yf
tsmc = yf.download('2330.TW', start='2023-01-01', end='2023-12-31')
```

```python
# 計算股價移動平均
sma5 = tsmc.Close.rolling(window=5).mean()
sma10 = tsmc.Close.rolling(window=10).mean()
```

```python
# 透過直條圖顯示
import matplotlib.pyplot as plt
plt.plot(tsmc.Close, "k--")
plt.plot(sma5, "r-.")
plt.plot(sma10, "b-.")
```

```python
# 透過yfinance抓取過去一年台積電股價
import yfinance as yf
data = yf.download('2330.TW', start='2023-01-01', end='2024-12-31')

# 計算股價移動平均
data['MA50'] = data['Adj Close'].rolling(window=50).mean()
data['MA200'] = data['Adj Close'].rolling(window=200).mean()

# 透過直條圖顯示
import matplotlib.pyplot as plt
plt.plot(data.index, data['Adj Close'], label='Adj Close')
plt.plot(data.index, data['MA50'], label='MA50')
plt.plot(data.index, data['MA200'], label='MA200')

```
